```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.style.use("fivethirtyeight")    

%matplotlib inline

#To read stock data from Yahoo!
from pandas_datareader.data import  DataReader #Install this module
import yfinance as yf

from pandas_datareader import data as pdr

yf.pdr_override()

#For time stamps
import datetime as datetime

#The tech stocks we weill use for analysis
tech_list= ['AAPL', 'GOOG', 'MSFT', 'AMZN', 'TSLA']

#Set up end and start times for the data grab
end = datetime.datetime.now()
# endtime=datetime.datetime.now()
start_time=datetime.datetime(end.year-1, end.month, end.day)

#To dispplay the data in a list.
for stock in tech_list:
    globals()[stock]= yf.download(stock, start_time, end)

company_list=[AAPL,GOOG, MSFT, AMZN, TSLA]
company_name= ['APPLE','GOOGLE', 'MICROSOFT', 'AMAZON', 'TSLA']


for company_name, com_name in zip(company_list, company_name):
    company_name["company_name"]=com_name
    
df=pd.concat(company_list,axis=0)
df.tail(10)

```

```python
import yfinance as yf
import pandas as pd

tech_list = ['AAPL', 'GOOG', 'MSFT', 'AMZN', 'TSLA']
start_time = '2023-01-01'  # Replace with your desired start date
end = '2023-12-31'  # Replace with your desired end date

company_data = {}

for stock in tech_list:
    data = yf.download(stock, start=start_time, end=end)
    company_data[stock] = data

company_name = ['APPLE', 'GOOGLE', 'MICROSOFT', 'AMAZON', 'TSLA']

for stock, com_name in zip(tech_list, company_name):
    company_data[stock]["company_name"] = com_name
 
df=pd.concat(company_list,axis=1)
df.tail(10)
   
```
